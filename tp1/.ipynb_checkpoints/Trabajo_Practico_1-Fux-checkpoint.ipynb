## Ejercicio 1 - Implementar clase PCA con numpy
**Dataset** Human activity recognition

Tomar las primeras 63 componentes principales y calcular la varianza contemplada. Realizar las operaciones internas con numpy.linalg. 

```python
## not implemented
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns
import os
import gzip
import sys
import pandas as pd
from sklearn.neighbors import KernelDensity
from sklearn.preprocessing import normalize 

PATH_DATASET = "data/UCI HAR Dataset/train"
X_train = np.loadtxt(os.path.join(PATH_DATASET, "X_train.txt"))
#y_train = np.loadtxt(os.path.join(PATH_DATASET, "y_train.txt")).astype(int)

#estructura
print("Estructura features del dataset: {}".format(X_train.shape))
print("Estructura de las etiquetas: {}".format(y_train.shape))

df = pd.DataFrame(data=X_train)

class PCA:
    
    def standarize(self, df):
        #obtengo la media y el desv√≠o standard para normalizar
        df_mean = df.mean()
        df_std = df.std()
        
        df = (df - df_mean) / df_std
        # retorno los datos normalizados
        return df
    
    def fit(self, components, df):       
        
        #guardo components
        self.components = components

        #normalize data 
        #data_norm = df - df.mean()
        #print("Datos normalizados:{}".format(data_norm.head()))
        self.df = self.standarize(df)

        #obtengo la matriz covarianza
        cov_mat = self.df.cov()
        
        #obtengo los autovalores y autovectores de la matriz de covarianza
        auto_valores, auto_vectores = np.linalg.eig(cov_mat)
        
        #print("eigvalues:shape={}".format(eig_values.shape))
        #print("e_indices:shape={}".format(np.argsort(eig_values).shape))
        #e_indices = np.argsort(eig_values)[::-1]
        # ordeno los indices en orden creciente
        auto_indices = np.argsort(auto_valores)
        # ordeno los autovectores con los indices
        self.auto_vectores_sorted = auto_vectores[:,auto_indices]
        
        varianza_contemplada = []
        #prev_var = 0
        for i in auto_valores:
            #prev_var = prev_var + (i/sum(auto_valores)).astype(float)
            varianza_contemplada.append((i/sum(auto_valores)).astype(float))
        varianza_contemplada = np.cumsum(varianza_contemplada)    
            
        #print(varianza_contemplada[:63])
        
        return varianza_contemplada[:self.components]
        
    def transform(self):
        proyeccion = (self.auto_vectores_sorted[:self.components])
        #print(proyeccion)
        
        res = self.df.dot(proyeccion.T)
        #print("res:shape ={}".format(res.shape))
        return res.to_numpy()


pca = PCA()
varianza_contemplada = pca.fit(200, df)

plt.figure(figsize=(10,7))
plt.plot(varianza_contemplada, color='k', lw=2)
plt.xlabel('Cantidad de componentes')
plt.ylabel('Total varianza contemplada')
plt.xlim(0, 200)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.axvline(63, c='b')
plt.axhline(0.9, c='r')
plt.show();

varianza_contemplada2 = pca.fit(63, df)
X_pca = pca.transform()

#print("X_pca:col0:{}".format(X_pca[:, 0]))

# Graficamos los primeros dos componentes
plt.figure(figsize=(12,10))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_train, alpha=0.7, s=40, cmap=plt.cm.get_cmap('nipy_spectral', 6))
plt.rcParams['axes.grid'] = False #Evita un warning
plt.colorbar()
plt.title('HAR - PCA projection 2D');




```

## Ejercicio 2 - Implementar clase KMeans con numpy
**Dataset** Human activity recognition

Implementar kmeans con numpy. Agrupar el dataset transformado (ejercicio de PCA) y agrupar en clusters de k=2 y 6. Graficar los casos de k=2 y k=6 con las primeras dos componentes principales.

```python
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from sklearn.datasets import make_blobs

class KMeans:
    def __init__(self, n_clusters, n_init):
        self.n_clusters = n_clusters
        self.n_init = n_init

      
    
    
    
    
    def fit(self, X):
        
          # Generate dataset
        #X = make_blobs(centers=self.n_clusters, n_samples=self.n_init, random_state=1)

        # Visualize
        fig, ax = plt.subplots(figsize=(4,4))
        ax.scatter(X[:,0], X[:,1], alpha=0.5)
        ax.set_xlabel('$x_1$')
        ax.set_ylabel('$x_2$');
        
        # Initialize our centroids by picking random data points
        centroids = self.initialize_clusters(X, self.n_clusters)

        # Initialize the vectors in which we will store the
        # assigned classes of each data point and the
        # calculated distances from each centroid
        classes = np.zeros(X.shape[0], dtype=np.float64)
        distances = np.zeros([X.shape[0], self.n_clusters], dtype=np.float64)

        # Loop for the maximum number of iterations
        for i in range(self.n_init):

            # Assign all points to the nearest centroid
            for i, c in enumerate(centroids):
                distances[:, i] = self.get_distances(c, X)

            # Determine class membership of each point
            # by picking the closest centroid
            classes = np.argmin(distances, axis=1)

            # Update centroid location using the newly
            # assigned data point classes
            for c in range(self.n_clusters):
                centroids[c] = np.mean(X[classes == c], 0)

    
    def initialize_clusters(self, points, k):
        """Initializes clusters as k randomly selected points from points."""
        return points[np.random.randint(points.shape[0], size=k)]
    
    # Function for calculating the distance between centroids
    def get_distances(self, centroid, points):
        """Returns the distance the centroid is from each data point in points."""
        return np.linalg.norm(points - centroid, axis=1)

                           
kmeans = KMeans(n_clusters=6, n_init=100)
kmeans.fit(X_pca)
cluster_labels = ['k', 'y', 'g', 'b', 'r', 'w']
                           
# Graficamos los primeros dos componentes - clusters id
plt.figure(figsize=(12,10))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, edgecolor='none', alpha=0.7, s=40, cmap=plt.cm.get_cmap('nipy_spectral', 6))
plt.rcParams['axes.grid'] = False
plt.colorbar()
plt.title('HAR - PCA projection 2D');
```
